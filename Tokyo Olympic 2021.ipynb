The Olympic Games are considered the world's foremost sports competition with more than 200 nations participating. The Olympic Games are normally held every four years, alternating between the Summer and Winter Olympics every two years in the four-year period. In Tokyo Olympic,around 11,656 atheletes took part from 205+ Countries.** 

## Importing Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

##  Finding Insights

**Arrange Countries according to  their Rank**

```python
top10_rank_df=olympic_df.sort_values('World_rank',ascending=True).head(10)
top10_rank_df
```

#### USA Has Top ranking because it  has won highest number of medals

## Arrange Countries according to highest Gold Medals Wins

```python
top10_gold_df=olympic_df.sort_values('Gold',ascending=False).head(10)
top10_gold_df
```

## Arrange Countries according to highest Silver Medals Wins

```python
top10_silver_df=olympic_df.sort_values('Silver',ascending=False).head(10)
top10_silver_df
```

## Arrange Countries according to highest Bronze Medals Wins

```python
top10_bronze_df=olympic_df.sort_values('Bronze',ascending=False).head(10)
top10_bronze_df
```

## Arrange Countries according to highest Total Medals Wins

```python
top10_total_df=olympic_df.sort_values('Total',ascending=False).head(10)
top10_total_df
```

## Find the Gold,Silver,Bronze Medals Won by India. Also finds its Ranking.

```python
india_info=olympic_df.query('Country=="India"')
india_info
```

### Find the Gold,Silver,Bronze Medals Won by Norway. Also finds its Ranking.

```python
Norway_info=olympic_df.query('Country=="Norway"')
Norway_info
```

## Find Countries Which has Silver Medal equal to 10

```python
Silver_info=olympic_df.query('Silver==10')
Silver_info
```

## Find Countries Which has total Medal Greater than 45

```python
total_info=olympic_df.query('Total>=45')
total_info
```

## Data visualization


```python
sns.pairplot(olympic_df);
```

```python
plt.figure(figsize=(20,14))
sns.barplot(x='Country',y='Gold',data=olympic_df)
plt.plot(olympic_df.Country,olympic_df.Gold,marker='s',c='g',ls='-',lw=3,ms=5,mew=1,mec='navy');
plt.tick_params(axis='x',which='major',labelsize=13,rotation=90)
plt.tick_params(axis='y',which='major',labelsize=20)
plt.title("Gold Medal Won By Each Countries",fontsize=30,pad=20)
plt.xlabel("Countries",fontsize=25)
plt.ylabel("Gold Medals Won",fontsize=25)
plt.legend(['Countries',' Gold Medals Won']);
plt.grid(color='g',linestyle='-',linewidth=0.3,alpha=0.5)
```

```python
plt.figure(figsize=(20,14))
sns.barplot(x='Country',y='Bronze',data=olympic_df)
plt.plot(olympic_df.Country,olympic_df.Bronze,marker='s',c='g',ls='-',lw=3,ms=5,mew=1,mec='navy');
plt.tick_params(axis='x',which='major',labelsize=13,rotation=90)
plt.tick_params(axis='y',which='major',labelsize=20)
plt.title("Bronze Medal Won By Each Countries",fontsize=30,pad=20)
plt.xlabel("Countries",fontsize=25)
plt.ylabel("Bronze Medals Won",fontsize=25)
plt.legend(['Countries',' Bronze Medals Won']);
plt.grid(color='g',linestyle='-',linewidth=0.3,alpha=0.5)
```

```python
plt.figure(figsize=(20,14))
sns.barplot(x='Country',y='Silver',data=olympic_df)
plt.plot(olympic_df.Country,olympic_df.Silver,marker='s',c='y',ls='-',lw=3,ms=5,mew=1,mec='navy');
plt.tick_params(axis='x',which='major',labelsize=13,rotation=90)
plt.tick_params(axis='y',which='major',labelsize=20)
plt.title("Silver Medal Won By Each Countries",fontsize=30,pad=20)
plt.xlabel("Countries",fontsize=25)
plt.ylabel("Silver Medals Won",fontsize=25)
plt.legend(['Countries',' Silver Medals Won']);
plt.grid(color='g',linestyle='-',linewidth=0.3,alpha=0.5)
```

```python
plt.figure(figsize=(20,14))
sns.barplot(x='Country',y='World_rank',data=olympic_df)
plt.plot(olympic_df.Country,olympic_df.World_rank,marker='s',c='g',ls='-',lw=3,ms=5,mew=1,mec='navy');
plt.tick_params(axis='x',which='major',labelsize=13,rotation=90)
plt.tick_params(axis='y',which='major',labelsize=20)
plt.title("Ranking of Each Countries",fontsize=30,pad=20)
plt.xlabel("Countries",fontsize=25)
plt.ylabel("World Rank ",fontsize=25)

plt.grid(color='g',linestyle='-',linewidth=0.3,alpha=0.5)
```

```python
plt.figure(figsize=(20,14))
plt.plot(olympic_df.Country,olympic_df.Gold,marker='s',c='r',ls='-',lw=1,ms=4,mew=1,mec='navy');
plt.plot(olympic_df.Country,olympic_df.Silver,marker='s',c='b',ls='-',lw=1,ms=4,mew=1,mec='navy');
plt.plot(olympic_df.Country,olympic_df.Bronze,marker='s',c='g',ls='-',lw=1,ms=4,mew=1,mec='navy');

plt.tick_params(axis='x',which='major',labelsize=13,rotation=90)
plt.tick_params(axis='y',which='major',labelsize=20)
plt.title("Total Medals Won By Each Countries",fontsize=30,pad=20)
plt.xlabel("Countries",fontsize=25)
plt.ylabel(" Medals Count",fontsize=25)
plt.legend([' Gold Medals Won','Silver Medals Won','Bronze Medals Won']);
plt.grid(color='g',linestyle='-',linewidth=0.3,alpha=0.5)

```
